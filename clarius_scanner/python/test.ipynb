```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your data and reshape it
df = pd.read_csv("testData_192x1168.csv")
reshaped_df = df.values.reshape(192, 1168)
df = pd.DataFrame(reshaped_df)

# Extract the row you want to plot
row = df.iloc[61]

# Create a Matplotlib figure with a larger size (in inches)
plt.figure(figsize=(16, 6))  # Adjust the width and height as needed

# Plot the data for a specific x-axis range (e.g., from 300 to 600)
x_range = np.linspace(0, 1167, 1168)
plt.plot(x_range, row)
plt.xlim(300, 600)  # Set the x-axis limits

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')

# Display the plot
plt.show()
```
