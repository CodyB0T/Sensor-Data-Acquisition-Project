```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your data and reshape it
df = pd.read_csv("testData.csv")
reshaped_df = df.values.reshape(104, 592)
df = pd.DataFrame(reshaped_df)

# Extract the first row
row = df.iloc[61]

# Create a Matplotlib figure with a specific size (in inches)
plt.figure(figsize=(12, 6))  # Adjust the width and height as needed

# Plot the data
plt.plot(np.linspace(0, 591, 592), row)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')

# Display the plot
plt.show()
```
