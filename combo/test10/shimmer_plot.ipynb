## plot shimmer imu

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/shimmer_data.csv")
df

fig, ax = plt.subplots(3, 1, figsize=(20,10))

x=df.iloc[0].values
y=df.iloc[1].values
z=df.iloc[2].values


ax[0].plot(x)
ax[0].set_title("shimmer x")

ax[1].plot(y)
ax[1].set_title("shimmer y")

ax[2].plot((z - 2084)/367.5)
ax[2].set_title("shimmer z")

# 4096

plt.show()
```
