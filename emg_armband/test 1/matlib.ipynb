```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Loop through the rows and create individual plots
for index, row in df.iterrows():
    x_values = row['X']
    row_data = row[1:]  # Exclude the 'X' column

    plt.figure(figsize=(8, 5))  # Set the figure size for each plot
    plt.plot(x_values, row_data)
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.title(f'Plot for Row {index + 1}')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

# Create a single figure with subplots
fig, axes = plt.subplots(7, 1, figsize=(20, 14))  # Create 7 subplots in a single column

for row in range(1, 8):
    data = []
    for x in range(df.shape[1]):
        data.append(df.iloc[row, x])

    axes[row - 1].plot(data)  # Use the respective subplot for each row
    axes[row - 1].set_title(f"Channel {row}")

plt.tight_layout()  # Adjusts spacing between subplots
plt.show()
```
